```python
puzzle = Puzzle(MATCH_IMGS)
```

```python
corner_piece = puzzle.pieces[3]
corner_piece.insert(canvas)
corner_piece.inserted = True

queue = []
queue.append(corner_piece)
```

```python
while queue: # Keep looping over pieces as long as queue is not empty
    counter = 0 # counter to make sure that we stop bfs loop after four edges
    s = queue.pop(0) # remove top entry in the queue, this will be our first piece that we initialised with
    for edge in s.return_edge(): # check correspoinding edges of piece we popped from the queue
        counter += 1 # track that we have checked the first edge
        if edge != None and edge.connected_edge != None: # make sure that we are not working with flat edges
            for piece in puzzle.pieces: # loop through all puzzle pieces to track connecting edges
                if piece.inserted == False: # make sure we are not rechecking an already inserted piece
                    for puzzle_edge in piece.edge_list: # check all four edges of piece to be inserted
                        if puzzle_edge != None: # make sure we dont check flat edges
                            if (puzzle_edge.point1 == edge.connected_edge.parent_piece.top_edge.point1).all():
                                queue.append(piece) # enque inserted piece so that we can repeat process
                                puzzle.pieces[piece.idx].insert(canvas) # connect piece to the canvas
                                puzzle.pieces[piece.idx].inserted = True # mark inserted piece as connected
                                canvas.display_canvas()
        if counter > 4: # break bfs edge loop after checking four edges
            break
```
